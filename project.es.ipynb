#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
# COMPARACIÓN DE PRECIOS POR METRO CUADRADO

# Precios por metro cuadrado

ds['pps'] = ds['price'] / ds['surface']
ds['pps'] = ds['pps'].fillna(0)

# Precio promedio por metro cuadrado de las casas en Valdemorillo

pps_valdemorillo = ds.loc[ds['level5'] == 'Valdemorillo', 'pps']
pps_valdemorillo_promedio = pps_valdemorillo.mean()

print(f"Precio promedio por metro cuadrado de las casas de Valdemorillo: {pps_valdemorillo_promedio}")

# Precio promedio por metro cuadrado de las casas en Galapagar

pps_galapagar = ds.loc[ds['level5'] == 'Galapagar', 'pps']
pps_galapagar_promedio = pps_galapagar.mean()

print(f"Precio promedio por metro cuadrado de las casas de Galapagar: {pps_galapagar_promedio}")

# Análisis y comparación

if pps_valdemorillo_promedio > pps_galapagar_promedio:
    print("Por lo tanto, las casas en Valdemorillo son más caras por metro cuadrado")
else:
    print("Por lo tanto, las casas en Galapagar son más caras por metro cuadrado")

```

#### Ejercicio 11. Analiza la relación entre la superficie y el precio de las casas. (★★☆)

Pista: Puedes hacer un `scatter plot` y luego escribir una conclusión al respecto.

```python
# COMPARACIÓN DE PRECIOS POR SUPERFICIE

from sklearn.linear_model import LinearRegression
import numpy as np

# Gráfico de dispersión de precio vs superficie

plt.scatter(ds['surface'], ds['price'], color='blue')
plt.title('Precio vs Superficie')
plt.xlabel('Superficie')
plt.ylabel('Precio')

formatter = ticker.ScalarFormatter(useMathText=False)
formatter.set_scientific(False)
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()

# Regresión lineal

ds['surface'] = ds['surface'].fillna(0)
ds['price'] = ds['price'].fillna(0)
X = np.array(ds['surface']).reshape(-1, 1)
Y = np.array(ds['price'])
modelo = LinearRegression()
modelo.fit(X, Y)

# Coeficiente de correlación

r2 = modelo.score(X, Y)
print(f"Coeficiente de correlación: {r2}")

if r2 >= 0.9:
    print("La correlación es muy alta")
elif r2 >= 0.7:
    print("La correlación es alta")
elif r2 >= 0.5:
    print("La correlación es moderada")
elif r2 >= 0.3:
    print("La correlación es baja")
else:
    print("La correlación es muy baja")

print("Por lo tanto, el precio de las casas no depende de la superficie")
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
import folium
from IPython.display import display


coordenadas_poblaciones_interes = subconjunto.loc[subconjunto['level5'].isin(poblaciones_interes), ['latitude', 'longitude']]

# Conversión de las coordenadas a float

coordenadas_poblaciones_interes['latitude'] = coordenadas_poblaciones_interes['latitude'].str.replace(',', '.').astype(float)
coordenadas_poblaciones_interes['longitude'] = coordenadas_poblaciones_interes['longitude'].str.replace(',', '.').astype(float)

# Creación de un DataFrame con las coordenadas de las poblaciones de interés

ds_coordenadas_poblaciones_interes = {
    'poblacion' : subconjunto['level5'],
    'latitude': coordenadas_poblaciones_interes['latitude'],
    'longitude': coordenadas_poblaciones_interes['longitude']
}

ds_coordenadas_poblaciones_interes = pd.DataFrame(ds_coordenadas_poblaciones_interes)

# Creación de un mapa con las coordenadas de las poblaciones de interés

mapa = folium.Map(location=[40.416775, -3.703790], zoom_start=10)

for _, row in ds_coordenadas_poblaciones_interes.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['poblacion'],
        icon=folium.Icon(color='blue' if row['poblacion'] == 'Fuenlabrada' else
                        'green' if row['poblacion'] == 'Leganés' else
                        'red' if row['poblacion'] == 'Getafe' else
                        'purple')
).add_to(mapa)

print(ds_coordenadas_poblaciones_interes)
display(mapa)

#map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    #basemap=basemaps.Stamen.Terrain)
#map
```
